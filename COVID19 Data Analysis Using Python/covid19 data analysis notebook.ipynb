## 1. Importing the modules 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

## 2.  Dataset - I

```python
# Importing Covid-19 dataset
corona_dataset_csv = pd.read_csv('covid19_Confirmed_dataset.csv')
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
#266 rows and 104 cols
```

```python
# Delete the useless columns
corona_dataset_csv.drop(['Lat','Long'],axis=1,inplace=True)
corona_dataset_csv.head(10)
```

```python
# Aggregating the rows by the country
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.shape
```

```python
# Visualizing data related to a country for example China
# Cumulative number of confirmed cases
corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
corona_dataset_aggregated.loc['Spain'].plot()
plt.legend()
```

## 3. Calculating a good measure 
We need to find a good measure represented as a number, describing the spread of the virus in a country. 

```python
corona_dataset_aggregated.loc['China'][:3].plot()
# In 24hrs, 100 new infected cases were recorded from 550 to 650 and the next day, the growth of infected cases was +250
```

```python
# Calculating the first derivative of the curve
corona_dataset_aggregated.loc['China'].diff().plot()
# Shows us the change in infection rate day by day in our period.
# And what we are looking for is this maximum number.
```

```python
# Finding maxmimum infection rate for China
corona_dataset_aggregated.loc['China'].diff().max()
# 15,136 new infected cases has been recorded in only 24 hours!
```

```python
# Finding maxmimum infection rate for Italy
corona_dataset_aggregated.loc['Italy'].diff().max()
```

```python
# Finding maxmimum infection rate for Spain
corona_dataset_aggregated.loc['Spain'].diff().max()
```

```python
# Finding maximum infection rate for all of the countries 
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for country in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated['max infection rate'] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
# Creating a new dataframe with only needed column
corona_data = pd.DataFrame(corona_dataset_aggregated['max infection rate'])
#.sort_values('max infection rate', ascending=False)
corona_data.head()
```

## 4. Dataset - II
- Importing the WorldHappinessReport.csv dataset
- Selecting needed columns for our analysis 
- Joining the datasets 
- Calculating the correlations as the result of our analysis

### Importing the dataset

```python
world_happiness_report = pd.read_csv("worldwide_happiness_report.csv")
world_happiness_report.head()
```

```python
world_happiness_report.shape
```

```python
# Dropping the useless columns 
columns_to_dropped = ['Overall rank','Score','Generosity','Perceptions of corruption']
world_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)
```

```python
world_happiness_report.head()
```

```python
# Changing the indices of the dataframe
world_happiness_report.set_index(['Country or region'],inplace=True)
world_happiness_report.head()
```

### Joining the two prepared dataset

#### Corona Dataset :

```python
corona_data.head()
```

```python
corona_data.shape
```

#### World Happiness Report Dataset :

```python
world_happiness_report.head()
```

```python
world_happiness_report.shape
```

```python
data = corona_data.join(world_happiness_report, how="inner") #Performing inner join because the number of countries in corona data is more than happiness report
data.head()
```
